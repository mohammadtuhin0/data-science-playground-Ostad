```python
# Some basic plots Using seaborn

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
sns.set(style='white')

rs = np.random.RandomState(10)
d = rs.normal(size=100)

sns.histplot(d, kde=True, color='m')
```

```python
# Distplot

sns.set(style='white')

rs = np.random.RandomState(10)
data = rs.normal(size=100)

colors = ['r', 'g', 'b']

sns.distplot(data, kde=True, hist=True, bins=10,
            rug=True, hist_kws={"alpha":0.3,
                               "color": colors[0]},
            kde_kws={"color": colors[1], "lw": 2},
            rug_kws={"color": colors[2]})
```

```python
# Lineplot

sns.set(style='dark')
fmri = sns.load_dataset('fmri')

sns.lineplot(x='timepoint',
            y='signal',
            hue='region',
            style='event',
            data=fmri)
```

```python
# Lmplot

sns.set(style="ticks")

df = sns.load_dataset("anscombe")
sns.lmplot(x="x", y="y", data=df)
```

```python
# Seaborn.relplot()

# Example 1: Visualizing the most basic plot to show all the data points in tips dataset.

sns.set(style='ticks')

tips = sns.load_dataset('tips')

sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
# Example 2: Grouping data points on the basis of category, here as time.

sns.set(style='ticks')

tips = sns.load_dataset('tips')

sns.relplot(x='total_bill', y='tip', hue='time', data=tips)
```

```python
# Example 3: using time and sex for determining the facet of the grid.

sns.set(style='ticks')

tips = sns.load_dataset('tips')

sns.relplot(x='total_bill', y='tip', hue='day', col='time', row='sex', data=tips)
```

```python
# Example 4: using size attribute, we can see data points having different size.

sns.set(style='ticks')

tips = sns.load_dataset('tips')

sns.relplot(x='total_bill', y='tip', hue='day', size='size',)
```
